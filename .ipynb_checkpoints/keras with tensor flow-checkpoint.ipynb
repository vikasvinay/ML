```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels =[]
train_samples = []
```

```python
for i in range(50):
    
# 50 younger people exprienced side effects (5%)
    random_younger = randint(13,64) #age group of 13 to 64
    train_samples.append(random_younger)
    train_labels.append(1)

# 50 old people exprienced no side effects (5%)
    random_old = randint(65,100)
    train_samples.append(random_old)
    train_labels.append(0)
    
for i in range(1000):
# 1000 younger people exprienced no side effects(95%)
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
# 1000 old people exprienced side effects (95%)
    random_old = randint(65, 100)
    train_samples.append(random_old)
    train_labels.append(1)
    
```

```python
#convert list to array
train_samples = np.array(train_samples)
train_labels = np.array(train_labels)

#shuffle the data 
train_samples, train_labels = shuffle(train_samples, train_labels)
```

```python
# reshaping train_sample to range 0 to 1

scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_sample = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
scaled_train_sample
```

 # Simple tf.keras Sequential Model 

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='softmax')
])
```

```python
model.summary()
```

```python
# Prepares the model and arranges everything
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# validation_split, which splits some testing data into validation set
model.fit(x=scaled_train_sample, y=train_labels, validation_split=0.1,batch_size=10, epochs=30, shuffle=True, verbose=2)
```

# Confusion Matrix

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true=real_lable, y_pred=rounded_predictions)
```

```python
cm
```

```python
def plot_confusion_matrix(cm, classes,
                        normalize=False,
                        title='Confusion matrix',
                        cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects', 'had_side_effects']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels)
```
